```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def trimf(x,a,b,c):
    rs = np.nan
    if x<a or x>c:
        rs = 0
    elif x>=a and x<=b:
        rs = (x-a)/(b-a)
    elif x>b and x<=c:
        rs = (c-x)/(c-b)
    return rs

def solgrid(t,z,u,par):
    rs = np.zeros(shape=(len(z),len(t)))
    for i,vt in enumerate(t):
        for j,vz in enumerate(z):
            rs[j,i] = trimf(vz-vt+u*vt,par[0],par[1],par[2])
    return rs
```

```python
a = 0
beta = 2
e = 1
x0 = 5
T = 1
u = beta*e*(T+x0-a)/(1+beta*e*T)

t = np.linspace(0,T,500)
z = np.linspace(x0-e,0,500)
T, Z = np.meshgrid(t, z)
M = solgrid(t,z,u,[x0-e,x0,x0+e])

print(u)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(M, extent=[t.min(), t.max(), z.min(), z.max()],cmap='bone_r',aspect=1/8,origin='lower')
ax.set_xlabel('t')
ax.set_ylabel('z')
# cbar = fig.colorbar(im)
# cbar.set_label('p')

plt.show()
```
